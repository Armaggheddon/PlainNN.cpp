#🖼️ MNIST Handwritten Digit Classifier - Live Demo!
Welcome to the live demo of PlainNN.cpp! In this interactive demo, you’ll get to experience how a neural network trained on the MNIST dataset can recognize handwritten digits. 🎨✍️

## How it Works:
A simple webpage will open up with a canvas where you can draw a number (0-9) using your mouse or touchscreen.
Once you’ve drawn your digit, hit the "Classify Drawing" button to see the neural network's prediction in action! 🚀
The model will output scores for each class (0-9), showing which digit it thinks you’ve drawn, and how confident it is. 📊
No need to install anything—just draw, click, and watch the neural network do its magic. ✨
